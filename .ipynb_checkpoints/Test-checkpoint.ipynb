```python
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
import pandas as pd
```

```python
df = pd.read_csv('test.csv')
```

```python
df1 = df.copy()
```

```python
df1 = df1.dropna(subset = ["Model",'Color','ROM','Display size','Battery size','Processor'])
```

```python
df1.info()
```

```python
ram_map={'4GB': 4,
 '3GB': 3,
 '6GB': 6,
 '8GB': 8,
 '2GB': 2,
 '1GB': 1,
 '12GB': 12,
 '512MB': 1,
 '1.5GB': 2}
```

```python
df1['RAM'].replace(ram_map,inplace=True)
df1['RAM'].value_counts()
```

```python
ind = df1.loc[df1['ROM'].isin(['2018 Edition) ','2nd Generation) ','New 2016 Edition) ','With Turbo Charger) ','5G) ',' Gold','3rd Generation) ',' 1 TB','Pearl Black ',' 512 MB'])].index
```

```python
df1.drop(index=ind,inplace=True)
```

```python
rom_map = {' 128 GB': 128,
 ' 64 GB': 64,
 ' 32 GB': 32,
 ' 16 GB': 16,
 ' 256 GB': 256,
 ' 8 GB': 8,
 ' 4 GB': 4,
 ' 512 GB': 512}
```

```python
df1['ROM'].replace(rom_map,inplace=True)
```

```python
df1['ROM'].value_counts()
```

```python
df1.info()
```

```python
from sklearn.model_selection import train_test_split

X = df1[['RAM','ROM','Display size','Battery size']]
y = df1['MRP']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression

regressor = LinearRegression()
regressor.fit(X_train,y_train)
y_pred = regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
score = r2_score(y_test,y_pred)
score
```
